```python
from sklearn.datasets import  load_iris

import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
data=iris.data
target=iris.target

xx=[0.2,0.3,0.4,0.5]
accuracy_list=[]
for i in range(0,len(xx)):
    train_data, test_data, train_target_data, test_target_data = train_test_split(data,target,test_size=xx[i],shuffle=True)
    clf=tree.DecisionTreeClassifier()
    algo=clf.fit(train_data,train_target_data)
    output=algo.predict(test_data)
    accuracy=accuracy_score(test_target_data,output)
    accuracy_list.append(accuracy)
print(accuracy_list)
plt.scatter(xx,accuracy_list,label="bar graph",color='red',marker="x")
plt.plot(xx,accuracy_list,label="line equation",color='blue')
plt.xlabel("input data in percentage")
plt.ylabel("output data in percentage")
plt.legend()
plt.show()
```
