```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
dir(load_iris())
```

```python
iris=load_iris()

print(iris.data)

print(iris.feature_names)

print(iris.target_names)

#0--->setosa 1-->versicolor 2--> virginica
print(iris.target)

```

```python
# data shape
print(iris.data.shape)
```

```python
#index array for which we will test the data
x=[0,72,119]

# extracting training data and target data from data

training_data=np.delete(iris.data,x,axis=0)
training_target_data=np.delete(iris.target,x)
print(training_target_data)

# values of target and test of x array
targetData=iris.target[x]
testData=iris.data[x]
print(targetData)
```

```python
# calling Decison Tree Algorithm 
clf=tree.DecisionTreeClassifier(criterion="gini")
# passing data to algorithm to train data
training=clf.fit(training_data,training_target_data)
output=training.predict(testData)
print(output)
```

```python
plt.scatter(training_data[:,0],training_target_data,label="sepal length")
plt.scatter(training_data[:,1],training_target_data,label="sepal width")
plt.scatter(training_data[:,2],training_target_data,label="petal length")
plt.scatter(training_data[:,3],training_target_data,label="petal width")
plt.xlabel("x in cms")
plt.ylabel("output")
plt.title("feature v/s output")
plt.legend()
plt.show()
```

```python
# to draw tree 
import pydotplus
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
```

```python
dot_data=StringIO()
export_graphviz(clf,out_file=dot_data,feature_names=iris.feature_names,filled=True, 
                class_names=iris.target_names,rounded=True,special_characters=False)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
#accuracy calculation of tree
from sklearn import metrics as me
me.accuracy_score(targetData,output,normalize=True)
clf.score(training_data,training_target_data)
```
